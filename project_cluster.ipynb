```python
data['CREDIT_LIMIT'] = df['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean())
```

```python
data['MINIMUM_PAYMENTS'] = data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean())
```

```python
#scale the data
scaler = StandardScaler()
X= scaler.fit_transform(data)
```

```python
X.shape
```

```python
X
```

```python
#KMEANS
kmeans= KMeans(n_clusters = 5, random_state = 42)
kmeans.fit(X)

```

```python
pred = kmeans.predict(X)

pred
```

```python
#BEST CLUSTER FOR KMEANS
scores_1 = []

range_values = range(1, 20)

for i in range_values:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(X)
  scores_1.append(kmeans.inertia_) 
```

```python
plt.plot(scores_1, 'bx-')
plt.title('Finding the right number of cluster')
plt.xlabel('Cluster')
plt.ylabel('Scores')
plt.show()
```

```python
#so KMEANS with 8 cluster 
kmeans= KMeans(n_clusters = 8, random_state = 42)
kmeans.fit(X)
pred = kmeans.predict(X)

pred
```

```python
#cluster column in dataframe
data['Cluster'] = pd.DataFrame(pred, columns=['cluster'])
data
```

```python
data["Cluster"].unique()
```

```python
#plot hist cluster
for i in data.columns:
  plt.figure(figsize = (25, 5))
  for j in range(8):
    plt.subplot(1,8,j+1)
    cluster = data[data['Cluster'] == j]
    cluster[i].hist(bins = 20)
    plt.title('{}    \nCluster {} '.format(i,j))
  
  plt.show()
```
