Year and Month

```python
from datetime import datetime, timedelta
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df["years"]=df["order_date"].dt.year
df['month'] = df['order_date'].dt.month_name()
```

```python
sales_by_year=df.groupby("years").agg({"Sales":sum})
sales_by_month=df.groupby("month").agg({"Sales":sum})
```

```python
sales_by_year.plot(kind="bar",figsize=(10,5),title="sales_by_years")
```

```python
sales_by_month.plot(kind="bar",figsize=(10,5),title="sales_by_month")
```

```python
#orderday-shipday
df['shipment_days'] = df['ship_date'] - df['order_date']
df['shipment_days'] = df['shipment_days'].dt.days
df.head()
```

```python
Avg_shipmentday_by_month=df.groupby("month").agg({"shipmentdays":np.mean})
```
